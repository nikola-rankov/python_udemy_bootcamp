# Support Vector Machines

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
df_feat = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
```

## SVM classification

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_feat, np.ravel(df_target), test_size=0.30, random_state=101)
```

```python
model = SVC(gamma='auto')# decision_function_shape is default to 'ovr'. None is depreciated in version 0.17. Possible parameter values are 'ovr'(default) and 'ovo'
                        # gamma='scale' is the current default option
 
```

```python
model.fit(X_train,y_train)
```

```python
model.get_params()
```

```python
predictions = model.predict(X_test)
```

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```

# Gridsearch
## Optimise the tuning parameters of the SVM classifier using the inbuilt gridsearch method

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']}
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
print('\n')
print(classification_report(y_test,grid_predictions))
```
