```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

#### Functional Method

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

#### Object oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1 = fig.add_axes([0.1,0.5,0.4,0.3])
```

#### Part 2

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('X Label 1')
axes[0].set_ylabel('Y Label 1')
axes[0].set_title('Set Title 1')
axes[1].plot(y,x)
axes[1].set_xlabel('X Label 2')
axes[1].set_ylabel('Y Label 2')
axes[1].set_title('Set Title 2')
fig.show()
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
fig.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('my_picture.jpg',dpi=200)
```

#### Part 3:

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green')
ax.plot(x,y,color='#10CB10')# You can use RGF hex codes as well
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#10CB10',linewidth=2,alpha=1)# 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#10CB10',lw=2,linestyle='--')# linewidth == lw
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#10CB10',lw=2,ls='--',marker='o',markersize=10)# linestyle == ls
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#10CB10',lw=2,ls='--')# linestyle == ls
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
